```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dummy_insurance_data.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Age"],df["Insurance"],marker="*",color="red")
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df[["Age"]],df["Insurance"],test_size=0.1,random_state=0)
```

```python
X_train.nunique()
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.predict(X_test)
```

```python
y_pred = reg.predict(X_test)
```

```python
data =pd.DataFrame({'Actural value':y_test,'predicted value':y_pred})
data
```

```python
newdata = [[25]]

y_pred = reg.predict(newdata)
y_pred
```

```python
reg.score(X_test,y_test)
```
